```python
# Libs Necess치rias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para ana치lises gr치ficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')


url = 'https://raw.githubusercontent.com/marcofavero3/dfDados_Incendio/main/Dados_Incendio.csv'

df = pd.read_csv(url, encoding='latin-1')

df.head()

```

```python
# Nulos
df.isnull().sum()
```

```python
# Nulos Plot
plt.figure( figsize=(14,5) )
plt.title('An치lise de campos nulos')
sns.heatmap( df.isnull(), cbar=False );
```

```python
# Estatistica
df.describe()
```

```python
# Info
df.info()
```

```python
# Campos Unicos
df.nunique()
```

```python
# analise por ano dos incendios
Analise = df.groupby( by=['year'] ).sum().reset_index()

Analise.head()

# Tamanho
plt.figure( figsize=(12, 5) )

# Style
plt.style.use('ggplot')

# Grafico
plt.title( 'Total ind칡ncidios no Brasil: 1997 - 2017', loc='left', fontsize=14 )
sns.lineplot( data=Analise, x='year', y='number', estimator='sum', lw=2, color='#ff5555', alpha=0.85 );

# Labels
plt.xlabel('Quantidade')
plt.ylabel('Per칤odo');
```

```python
# analise por ano dos incendios
Analise_02 = df.groupby( by=['year', 'month'] ).sum().reset_index()
Analise_02.head()

# Tamanho
plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Ind칡ncidios por m칡s', loc='left', fontsize=14 )
sns.boxplot( data=Analise_02, x='month', y='number', palette='coolwarm', saturation=1, width=0.9, linewidth=2,
            order=['Janeiro', 'Fevereiro', 'Mar칞o', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'] )

# Labels
plt.xlabel('M칡s')
plt.ylabel('N칰mero de inc칡ndios');
```

```python
# analise por ano dos incendios
Analise_03 = df.groupby( by=['state'] ).sum().reset_index()[['state', 'number']].sort_values( 'number', ascending=False )
Analise_03.head()

# Tamanho
plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Estados com maior n칰mero de inc칡ndios', loc='left', fontsize=14 )

# Grafico
plt.bar( Analise_03.state, Analise_03['number'], color='#f44e3f' )

# Labels
plt.ylabel('Quantidade')
plt.xticks( rotation=90 );
```

```python
# Estados TOP 10
Lista_TOP10 = Analise_03['state'][0:10].values

# Tamanho
plt.figure( figsize=(12, 5) )

# Grafico
plt.title( 'Top 10 Estados com inc칡ncios', loc='left', fontsize=14 )

# Loop
for Coluna in Lista_TOP10:

  # Filtrar o estado
  Filtro = df.loc[ df['state'] == Coluna ]

  # Agrupar os valores para sumarizar
  Analise_Local = Filtro.groupby( by=['year'] ).sum().reset_index()

  # Plot
  sns.lineplot(data=Analise_Local, x='year', y='number', lw=2, alpha=0.85)

# Labels
plt.xlabel('Per칤odo')
plt.ylabel('N칰mero de inc칡ndios')

# Legenda
plt.legend( Lista_TOP10, bbox_to_anchor=(1, 0.7) );
```
